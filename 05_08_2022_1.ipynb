```python
import numpy as np
def binning(col_name):
    min_value=col_name.min()
    max_value=col_name.max()
    bins = np.linspace(min_value,max_value,5)
    labels = [0,1,2,3]
    col_name=pd.cut(col_name,bins=bins, labels=labels, include_lowest=True)
    return col_name
```

```python
col_names=["mean_CTA_len","product","category","subject_len","sender","is_price","mean_paragraph_len","body_len","no_of_CTA"]
```
