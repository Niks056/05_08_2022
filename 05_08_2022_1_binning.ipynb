```python
import pandas as pd
```

```python
train=pd.read_csv("train_F3fUq2S.csv")
test=pd.read_csv("test_Bk2wfZ3.csv")
```

```python
pd.set_option('display.max_column',100)
```

```python
train_updated=pd.get_dummies(train,drop_first=True)
test_updated=pd.get_dummies(test,drop_first=True)
```

```python
col_names=["mean_CTA_len","product","category","subject_len","sender","is_price","mean_paragraph_len","body_len","no_of_CTA"]
train_updated["mean_CTA_len"] = pd.cut(train_updated["mean_CTA_len"], 5, labels=[0,1,2,3,4])
train_updated["product"] = pd.cut(train_updated["product"], 5, labels=[0,1,2,3,4])
train_updated["category"] = pd.cut(train_updated["category"], 5, labels=[0,1,2,3,4])
train_updated["subject_len"] = pd.cut(train_updated["subject_len"], 5, labels=[0,1,2,3,4])
train_updated["sender"] = pd.cut(train_updated["sender"], 5, labels=[0,1,2,3,4])
train_updated["is_price"] = pd.cut(train_updated["is_price"], 5, labels=[0,1,2,3,4])
train_updated["mean_paragraph_len"] = pd.cut(train_updated["mean_paragraph_len"], 5, labels=[0,1,2,3,4])
train_updated["body_len"] = pd.cut(train_updated["body_len"], 5, labels=[0,1,2,3,4])
train_updated["no_of_CTA"] = pd.cut(train_updated["no_of_CTA"], 5, labels=[0,1,2,3,4])
train_updated[col_names]= train_updated[col_names].apply(pd.to_numeric)
```

```python
test_updated["product"] = pd.cut(test_updated["product"], 5, labels=[0,1,2,3,4])
test_updated["category"] = pd.cut(test_updated["category"], 5, labels=[0,1,2,3,4])
test_updated["subject_len"] = pd.cut(test_updated["subject_len"], 5, labels=[0,1,2,3,4])
test_updated["sender"] = pd.cut(test_updated["sender"], 5, labels=[0,1,2,3,4])
test_updated["is_price"] = pd.cut(test_updated["is_price"], 5, labels=[0,1,2,3,4])
test_updated["mean_paragraph_len"] = pd.cut(test_updated["mean_paragraph_len"], 5, labels=[0,1,2,3,4])
test_updated["body_len"] = pd.cut(test_updated["body_len"], 5, labels=[0,1,2,3,4])
test_updated["no_of_CTA"] = pd.cut(test_updated["no_of_CTA"], 5, labels=[0,1,2,3,4])
test_updated[col_names]= test_updated[col_names].apply(pd.to_numeric)
```

```python
test_updated_X=test_updated.drop(columns="campaign_id",axis=1)
X = train_updated.drop(columns=["campaign_id","click_rate"],axis=1)
y = train_updated['click_rate']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.70, random_state=18)
```

```python
X_1=pd.concat([X,X*18,(X*18)**2],axis=1)
test_updated_X=pd.concat([test_updated_X,test_updated_X*18,(test_updated_X*18)**2],axis=1)
```

```python
col_list_2=[]
col_list_3=[]
for x in X.columns:
    col_list_2.append(x+"_1")
    col_list_3.append(x+"_2")
```

```python
res_list = [*list(X.columns), *col_list_2,*col_list_3]
X_1.columns = res_list
test_updated_X.columns=res_list
```

```python
from catboost import CatBoostRegressor

cbr=CatBoostRegressor()
cbr.fit(X_train,y_train)
```

```python
cbr_val=cbr.predict(test_updated_X)
```
